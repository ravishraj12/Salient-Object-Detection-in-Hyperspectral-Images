```python
def corner_prior(img1, xcord, ycord):
	img_lab = img1
	sal_cent=[]
	length = img_lab.shape[1]
	breadth = img_lab.shape[0]
	cent = img_lab[xcord][ycord]
	for i in range((breadth)):
		for j in range((length)):
			sal_cent.append(scipy.spatial.distance.euclidean(img_lab[i][j][:], cent))
			# print('sal_cent', sal_cent)
	return sal_cent	
```

```python
def center_prior(img1):
	img1_rgb = np.float32(img1)
	img_lab = img1_rgb
	sal_cent=[]
	length = img_lab.shape[1]
	breadth = img_lab.shape[0]
	cent = img_lab[length//2][breadth//2]
	for i in range((breadth)):
		for j in range((length)):
			sal_cent.append(scipy.spatial.distance.euclidean(img_lab[i][j][:], cent))
	return sal_cent	
```

```python
def compute_weights(feature_maps):
    num_maps = len(feature_maps)
    weights = [0] * num_maps
    total_pixels = sum([fm.sum() for fm in feature_maps])
    for i in range(num_maps):
        if feature_maps[i].sum() > 0:
            weights[i] = feature_maps[i].sum() / total_pixels
    return weights
```

```python
def wavelet_contrast_prior(img):
    # Convert the image to grayscale if it's not already
    if img.ndim == 3:
        img = np.mean(img, axis=2)
    
    # Perform 2D wavelet decomposition using the Haar wavelet
    coeffs = pywt.dwt2(img, 'haar')
    cA, (cH, cV, cD) = coeffs

    # Compute the horizontal and vertical absolute differences of the coefficients
    diff_h = np.abs(cH)
    diff_v = np.abs(cV)

    # Compute the local mean and standard deviation of the coefficient differences
    mu_h = np.mean(diff_h)
    mu_v = np.mean(diff_v)
    std_h = np.std(diff_h)
    std_v = np.std(diff_v)

    # Compute the wavelet-based contrast prior map
    cp_h = (diff_h - mu_h) / (std_h + 1e-8)
    cp_v = (diff_v - mu_v) / (std_v + 1e-8)
    cp = np.sqrt(cp_h ** 2 + cp_v ** 2)

    # Upsample the contrast prior map to the original image size
    cp = cv2.resize(cp, (img.shape[1], img.shape[0]))

    # Normalize the contrast prior map to the range [0, 1]
    cp = (cp - np.min(cp)) / (np.max(cp) - np.min(cp))

    return cp
```

```python
def energy(I, I1, I2, I3, I4):
    return np.sum((I - I1)**2 + (I - I2)**2 + (I - I3)**2 + (I - I4)**2)
```

```python
def gradient(I, I1, I2, I3, I4):
    return 2 * (I - I1) + 2 * (I - I2) + 2 * (I - I3) + 2 * (I - I4)
```

```python
def optimize(I, I1, I2, I3, I4, alpha=0.1, max_iter=1000, threshold=1e-6):
    for i in range(max_iter):
        E_prev = energy(I, I1, I2, I3, I4)
        grad = gradient(I, I1, I2, I3, I4)
        I -= alpha * grad
        E_curr = energy(I, I1, I2, I3, I4)
        if np.abs(E_curr - E_prev) < threshold:
            break
    return I
```
